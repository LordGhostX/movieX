```python
import requests
from bs4 import BeautifulSoup
```

```python
series = "wolf"
r = requests.get(f"https://fztvseries.mobi/search.php?search={series}&beginsearch=Search&vsearch=&by=series")
page = BeautifulSoup(r.text, "html.parser")
```

```python
movies = page.find_all("div", {"class": "mainbox"})
for movie in movies:
    print("https://fztvseries.mobi/" + movie.find("a")["href"])
    print(movie.find("a").find("div")["id"], "\n")
```

```python
r = requests.get("https://fztvseries.mobi/subfolder-Airwolf.htm")
page = BeautifulSoup(r.text, "html.parser")
```

```python
seasons = page.find_all("div", {"class": "mainbox3"})
for season in seasons:
    print("https://fztvseries.mobi/" + season.find("a")["href"])
    print(season.find("a").text, "\n")
```

```python
r = requests.get("https://fztvseries.mobi/files-Airwolf--0.htm")
page = BeautifulSoup(r.text, "html.parser")
```

```python
episodes = page.find_all("div", {"class": "mainbox"})
for episode in episodes:
    print("https://fztvseries.mobi/" + episode.find("a")["href"])
    print(episode.find("a").find("div")["id"], "\n")
```

```python
download_session = requests.Session()
r = download_session.get("https://fztvseries.mobi/episode.php?fileid=21449")
page = BeautifulSoup(r.text, "html.parser")
pre_download_url = "https://fztvseries.mobi/download.php" + page.find("a", {"id": "dlink2"})["onclick"].split("download.php")[-1].split('";')[0]

r = download_session.get(pre_download_url)
page = BeautifulSoup(r.text, "html.parser")

links = page.find_all("input", {"name": "filelink"})
for link in links:
    print(link["value"], "\n")
```
